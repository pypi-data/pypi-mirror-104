```python
import light_curve_pure as lcp
import h5py
import matplotlib.pyplot as plt
```

```python
with h5py.File('testdata.hdf5', 'r') as f:
    data = f['dataset'][:]
```

```python
'''Amplitude'''
ampl = []

for i in range(len(data)):
    feature = lcp.Amplitude()
    if data[i][1].size == 0:
        continue
    ampl.append(feature(data[i][1], data[i][2], data[i][3]))
```

```python
plt.hist(ampl, bins=20)
plt.show()
```
